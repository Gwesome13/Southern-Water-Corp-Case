# Part 2: <span style="color:green">Data Cleansing</span>
I have shared my initial thoughts and concerns about the data set, and now it's time to clean and analyze it. At first glance, the data set appears to be fairly clean; however, there are a few aesthetic adjustments Iâ€™d like to make before diving into the analysis.

```python
print(jobs_in_data)
```

```python
jid = pd.DataFrame(jobs_in_data)
jid.columns = jid.columns.str.replace('_', ' ', regex=False)
print(jid)
```

```python
jid.columns = jid.columns.str.title()
print(jid)
```

# Part 2: (Step 2) <span style="color:green">Data Cleansing: Continued</span>
After cleaning the data, I confirmed my initial thoughts: it was already relatively clean. However, as shown in Row 8839 of the preview above, there is still some incomplete data in the dataset. Let's remove all incomplete data and ensure that 'USD' is fully capitalized in the "Salary in USD" column.

```python
jid = jid.dropna()
print(jid)
```

```python
jid.columns = [col.replace('Usd', 'USD') for col in jid.columns]
print(jid)
```

```python
if 'Salary' in jid.columns:
    jid['Salary'] = pd.to_numeric(jid['Salary'], errors='coerce')
    jid['Salary'] = jid['Salary'].apply(lambda x: f"{x:,}" if pd.notnull(x) else x) 

print(jid)
```

```python
if 'Salary In USD' in jid.columns:
    jid['Salary In USD'] = pd.to_numeric(jid['Salary In USD'], errors='coerce')
    jid['Salary In USD'] = jid['Salary In USD'].apply(lambda x: f"{x:,}" if pd.notnull(x) else x) 

print(jid)
```
